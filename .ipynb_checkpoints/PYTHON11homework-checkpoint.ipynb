### Задание 1

```python
import pandas as pd
```

```python
def filmclass(param):
    if param['rating']<=2:
        return 'Низкий рейтинг'
    elif param['rating']>2 and param['rating']<4.5:
        return 'Средний рейтинг'
    elif param['rating']>=4.5:
        return 'Высокий рейтинг'
    
stats = pd.read_csv('ratings.csv')
stats['class'] = stats.apply(filmclass, axis=1)
stats.head()
```

### Задание 2

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
file = pd.read_csv('keywords.csv')

def addregion(param):
    for i in geo_data['Центр']:
        if geo_data['Центр'][0] in param['keyword']:
            return 'Центр'
    for i in geo_data['Северо-Запад']:
        if geo_data['Северо-Запад'][0] in param['keyword']:
            return 'Северо-Запад'
    for i in geo_data['Дальний Восток']:
        if geo_data['Дальний Восток'][0] in param['keyword']:
            return 'Дальний Восток'
        else:
            return 'undefined'
        
file['region'] = file.apply(addregion, axis=1)

file[100:150]
```
